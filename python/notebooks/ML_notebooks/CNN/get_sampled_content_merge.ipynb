```python
labelled_level2 = pd.read_csv(os.path.join(DATADIR, 'labelled_level2.csv'), dtype=object)
#COLLAPSE World level2taxons
labelled_level2.loc[labelled_level2['level1taxon'] == 'World', 'level2taxon'] = 'world_level1'

#creating categorical variable for level2taxons from values
labelled_level2['level2taxon'] = labelled_level2['level2taxon'].astype('category')

#Get the category numeric values (codes) and avoid zero-indexing
labels = labelled_level2['level2taxon'].cat.codes + 1

#create dictionary of taxon category code to string label for use in model evaluation
labels_index = dict(zip((labels), labelled_level2['level2taxon']))
```

```python
# zero-indexing
labels0 = labelled_level2['level2taxon'].cat.codes 

#create dictionary of taxon category code to string label for use in model evaluation
labels_index0 = dict(zip((labels0), labelled_level2['level2taxon']))
```

```python
index1 = pd.DataFrame.from_dict(labels_index, orient='index')
index1['plus1'] = index1.index 
index1
index1.columns = ['index1taxon', 'plus1']
index1
```

```python
index0 = pd.DataFrame.from_dict(labels_index0, orient='index')
```

```python
index0['plus1'] = index0.index
index0.columns = ['index0taxon', 'plus1']
```

```python
merge_dict = pd.merge(
    index0, 
    index1, 
    on='plus1',
    how='outer',
    indicator=True
)
```

```python
merge_dict
```

```python
#create dictionary of taxon category code to string label for use in model evaluation
dict0to1 = dict(zip(merge_dict['index1taxon'], merge_dict['index0taxon']))
```

```python
random_sample['updated_level2taxon_correct'] = random_sample['level2taxon'].map(dict0to1).copy(deep=True)
```
