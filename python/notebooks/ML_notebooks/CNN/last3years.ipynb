```python
import pandas as pd
import os
import numpy as np
import altair as alt
```

```python
DATADIR = os.getenv('DATADIR')
df = pd.read_csv(os.path.join('/data', 'last_3_years.tsv'), sep='\t')
df
```

```python
chart = alt.Chart(df).mark_point().encode(
    alt.X('dev_support', title='Number of content items in dev set'),
    alt.Y('dev_f1', title='Performance (F1 score)'),
    tooltip=['taxon_label:N', 'dev_support:Q', 'dev_f1:Q']
).properties(
title='Level 2 model, only using data in last 3 years, data=10/7/18')
chart.save('recent_results_3years.html')
```
