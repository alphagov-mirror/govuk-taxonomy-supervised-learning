```python
import numpy as np
import pandas as pd
```

```python
taxons = pd.read_csv('../../data/clean_taxons.csv')
```

```python
content = pd.read_csv('../../data/clean_content.csv')
```

```python
taxons = taxons[['base_path','content_id','taxon_name','level1taxon','level2taxon','level3taxon','level4taxon']]
```

```python
content_taxons = pd.merge(
    left=content, 
    right=taxons, 
    left_on='taxon_id', 
    right_on='content_id', 
    how='outer', 
    indicator=True
)
```
